```python
from scipy.io import loadmat
import pandas as pd
import numpy as np
```

```python
file_path = '../Data/'
```

```python
## target =  'pred'
def mat_to_lst(target):
    res = loadmat('matlab/'+target+'.mat')
    detect_n = len(res[target][0])

    direction_lst = []
    event_idx_lst = []
    event_Timestamp_lst = []
    duration_lst = []

    for i in range(detect_n):
        src = res[target][0][i]
        direction_lst.append(src[0][0][0]) ##left 1 right 2 
        event_idx_lst.append(src[3][0])
        event_Timestamp_lst.append(src[5][0])
        duration_lst.append(src[6][0][0])
    
    tot_lst= [direction_lst, event_idx_lst, event_Timestamp_lst, duration_lst]

    return tot_lst
```

```python
pred  = mat_to_lst('pred')
```

```python
df = pd.read_csv(file_path + 'data_10Hz.csv')
df
```

```python
## make lc category for df (ans, res, pred)
df['pred'] = np.nan
df.fillna(0, inplace= True)

for event_num, event_idx in enumerate(pred[1]):
    direction = pred[0][event_num]
    df.iloc[event_idx,-1] = direction

df
```

```python
df.to_csv(file_path + 'data_10Hz_candidates.csv')
```
